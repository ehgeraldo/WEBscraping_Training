# IMPORTS

```python
from bs4 import BeautifulSoup

import requests
```

# WEBSCRAPING 

```python
# URL que desejamos informações
url = 'https://www2.hm.com/en_us/men/products/jeans.html'

# Dicionario => para completar a requisição ao website
headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}

# Efetua a requisição ao site desejado
page = requests.get( url, headers=headers )
```

```python
soup = BeautifulSoup( page.text, 'html.parser')
```

```python
# Todo html da página 
soup
```

```python
# encontra o primeiro registro (elemento)

products = soup.find( 'ul', class_='products-listing small' )
```

```python
products
```

```python
product_list = products.find_all( 'article', class_='hm-product-item')
product_list

```

```python
type( product_list )
```

```python
# código único para os produtos 

product_list[1].get( 'data-articlecode' )
product_list[2].get( 'data-articlecode' )
```

```python
# exemplo de aplicação de um for em uma linha
[ i*10 for i in range(1,10) ]
```

```python
# Aplicar um for para toda a lista

product_id = [ p.get( 'data-articlecode' ) for p in product_list ]
```

```python
# Retorna os id dos produtos da lista ( data-articlecode )

product_id
```

```python
product_category = [ p.get( 'data-category' )for p in product_list ]
```

```python
# Retorna a categoria da lista (data-category)
product_category
```

```python
products_list = products.find_all( 'a', class_='link')
```

```python
products_list
```

```python
# Para extrair os nomes dos produtos da lista

products_name = [ p.get_text() for p in products_list ]
```
