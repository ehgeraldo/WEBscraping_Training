# IMPORTS

```python

import requests

import numpy as np

from IPython.display import Image

from bs4 import BeautifulSoup
```

# WEBSCRAPING 

```python
Image( 'image_00.png' )
```

```python
# URL que desejamos informações
url = 'https://www2.hm.com/en_us/men/products/jeans.html'

# Dicionario => para completar a requisição ao website
headers = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5)AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}

# Efetua a requisição ao site desejado
page = requests.get( url, headers=headers )
```

```python
# Obtendo o html
page.text
```

```python
soup = BeautifulSoup( page.text, 'html.parser')
```

```python
Image( 'image_04.png' )
```

```python
soup
```

```python
Image( 'image_06.png' )
```

```python
# burcar o número de produtos

soup.find_all( 'h2', class_='load-more-heading' )
```

```python
total_item = soup.find_all( 'h2', class_='load-more-heading' )[0].get( 'data-total' )

total_item
```

```python
# Divide o número total de itens pela quantidade de produtos mostrados no website

page_number = np.round( int( total_item ) / 36 )
```

```python
# Concatenado

url02 = url + '?page-size=' + str( int( page_number*36 ) )
url02
```
